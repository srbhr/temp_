```python
import PyPDF2
```

```python
import os
```

```python
path = os.getcwd()
```

```python
def name_files(path):
    for (name, dircs, files) in os.walk(path+str("//PDF_Files")):
        return files
```

```python
resumes = name_files(path)
```

```python
for resume in resumes:
    paths = "PDF_Files//"+str(resume)    
    pdfFileObj = open(paths, 'rb')
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj)  
    pageObj = pdfReader.getPage(0) 
    print(pageObj.extractText())  
    pdfFileObj.close() 
    print("#"*100)
```
