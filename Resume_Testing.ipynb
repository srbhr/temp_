```python
import spacy
import pandas as pd 
```

```python
nlp = spacy.load("en_core_web_lg")
```

```python
df = pd.read_json('Dataset/Resume_1.json')
```

```python
data_list = []
```

```python
# This iterates over the enitities of the dataset.
for a in df.columns:
    data_list.append(df[a][0])
```

```python
rest = nlp(data_list[0])
```

```python
text = ""
lists = []
for token in rest:
    if token.is_stop == False:
        if token.pos_ in ['PROPN','NOUN','VERB']:
            text += token.lemma_
            text += " "
            lists.append(token.lemma_)
```
